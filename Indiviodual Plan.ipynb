# Data description

## Players

#### It has 196 obervations and 7 variables

+ `experience`: character, experience level of players, such as: `Pro`,`Veteran`,`Amateur`,`Regular`
+ `subscribe`: logical, whether or not player has subscribed, such as: `True` and `False`
+ `hashedEmail`: character, hashed email addrsses of players
+ `played_hours`: double, the game playing in hours 
+ `name`: character, names of players
+ `gender`: character, gender of players, as: `Male`,`Female`,`Non-binary`.`Prefer not to say`,`Agender`,`Two-SpiritedandOther`
+ `Age`: double, age of players

## Sessions

#### It has 1535 obervations and 5 variables

+ `hashedEmail`: character, hashed email addrsses of players
+ `start_time`: character, start time of the lastest game
+ `end_time`: character, end time of the lastest game
+ `original_start_time`: double, original data of start_time, not human readable
+ `original_end_time`: double, original data of end_time, not human readable

## Potential Issues

+ `HashedEmail` is a little bit useless in the data
+ The number of `Non-binary`, `Prefer not to say Agender`, `Two-Spirited` and `Other`are really small compare to `male` and `female`.


# Data Visualization

```r
players |> filter(played_hours <=10 & Age<= 50) |>
ggplot(aes(x=played_hours,y=Age))+
geom_point(aes(color=subscribe))+
labs(x="Played Time(Hours)",y="Age of players",color="Subscribe or not",
     title="Subscription for different age and played time of players")+
theme(text=element_text(size=15))
```

When we focus on players ≤ 50 years old who’ve played ≤ 10 hours.Those aged 10–30 who’ve played ≤ 4 hours are the most likely to subscribe, so we’ll use age and hours played as our predictors.

```r
ggplot(players, aes(x=subscribe))+ geom_bar() + facet_grid(cols = vars(experience)) 
ggplot(players, aes(x=subscribe))+ geom_bar(aes(fill=experience),position="dodge")
```

Amateurs subscribe most, Pros least; the rest are alike, so this predictor is weak.

```r
players |> group_by(gender) |>
summarize(max = max(played_hours),min = min(played_hours),avg = mean(played_hours))
```

```r
Gender has a very weak ability to distinguish the depth of play, and therefore should not be used as a primary predictor variable.
```
